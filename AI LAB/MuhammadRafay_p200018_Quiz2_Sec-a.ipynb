```python
np.random.seed(42)

x_values = np.random.uniform(low = 10 ,high = 100,size =500)

y_values = x_values + np.random.normal(loc = 0 , scale = 10 ,size =500)

plt.scatter(x_values,y_values)


plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plt of random values x and y ")

plt.xlim(10,100)
plt.ylim(0,200)


plt.show()

```
